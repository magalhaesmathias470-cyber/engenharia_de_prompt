```python
import time
import random

# Configuração
LIMITE_CRITICO = 80  # valor crítico

def ler_temperatura():
    """Simula leitura de temperatura"""
    return random.randint(50, 100)

def enviar_alerta(valor):
    """Simula envio de alerta (console)"""
    print(f"\n🚨 ALERTA CRÍTICO! Temperatura atingiu {valor}°C\n")

def monitorar():
    print("Monitoramento iniciado...\n")

    while True:
        temp = ler_temperatura()
        print(f"Temperatura atual: {temp}°C")

        if temp >= LIMITE_CRITICO:
            enviar_alerta(temp)

        time.sleep(2)  # espera 2 segundos

# Executar
monitorar()
```
